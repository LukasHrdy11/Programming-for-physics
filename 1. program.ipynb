# Medián, aritmetický a geometrický průměr

```python
from numpy import median, mean as arithmetic_mean
import numpy as np


def geometric_mean(a):
    a = np.asarray(a)
    if a.size == 0:
        return np.nan
    if np.any(a <= 0):
        raise ValueError("Geometric mean is defined for positive values only.")
    return float(np.exp(np.mean(np.log(a))))
```

```python
# vygeneruje náhodnou množinu 30 čísel (float v rozsahu 0-100)
rng = np.random.default_rng()
random_set = rng.random(30) * 100
rng = np.random.default_rng(42)
random_set = rng.random(30) * 100
print(random_set)
print("Mean:", arithmetic_mean(random_set))
print("Median:", median(random_set))
print("Geometric Mean:", geometric_mean(random_set))
```

```python
def aritmetic_median_geometricmean(data):
    """Vrátí aritmetický průměr, medián a geometrický průměr z dat."""
    return (arithmetic_mean(data), median(data), geometric_mean(data))
```

# Goniometrické funkce

```python
def function_sinus(x):
    return np.sin(x) + np.sin(11*x)/2 + np.sin(24*x)/3

def function_cosinus(x):
    return np.cos(x) + np.cos(11*x)/2 + np.cos(24*x)/3
```

```python
import matplotlib.pyplot as plt

# vykreslení existujících funkcí na existujícím vektoru n1
n1 = np.linspace(0, 2*np.pi, 1000)
y_sin = function_sinus(n1)
y_cos = function_cosinus(n1)

plt.figure(figsize=(10, 5))
plt.plot(y_cos, y_sin, label='složení funkcí kde čas je parametr', linewidth=2)
plt.axhline(0, color='gray', linewidth=0.7)
plt.xlabel('sinus')
plt.ylabel('cosinus')
plt.title('Skládané goniometrické funkce')
plt.legend()
plt.grid(True)
ax = plt.gca()
# spočítat délky jednotlivých úseků mezi po sobě jdoucími body a jejich součet
pts = np.column_stack((y_cos, y_sin))
diffs = np.diff(pts, axis=0)
seg_lengths = np.hypot(diffs[:, 0], diffs[:, 1])
total_length = seg_lengths.sum()
#print("Délky jednotlivých úseků:", seg_lengths)
#print("Součet délek:", total_length)
ax.text(0.95, 0.95, f"Součet délek: {total_length:.3f}", transform=ax.transAxes,
    ha='right', va='top', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))
ax.set_aspect('equal', adjustable='box')
plt.tight_layout()
plt.show()
```

```python
def test_goniometric(n):
    n1 = np.linspace(0, 2*np.pi, n)
    y_sin = function_sinus(n1)
    y_cos = function_cosinus(n1)
    pts = np.column_stack((y_cos, y_sin))
    diffs = np.diff(pts, axis=0)
    seg_lengths = np.hypot(diffs[:, 0], diffs[:, 1])
    total_length = seg_lengths.sum()
    return total_length

def plot_goniometric_dependency(n):
    # pokud je n skalár, vytvoříme posloupnost od 3 do n (nutné mít alespoň 3 body)
    ns = np.arange(3, n + 1)
    # spočítat součet délek pro každé n pomocí existující funkce test_goniometric
    lengths = [test_goniometric(int(k)) for k in ns]
    plt.figure(figsize=(8, 4))
    plt.plot(ns, lengths, marker='o', linestyle='-')
    plt.xlabel('počet bodů n')
    plt.ylabel('součet délek')
    plt.title('Závislost součtu délek křivky na počtu vzorků')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    return ns, np.array(lengths)

plot_goniometric_dependency(1000)
```
